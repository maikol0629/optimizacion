```python
import numpy as np

def golden_section_search(f, a, b, tol=1e-5):
    """
    Realiza la búsqueda de la razón dorada para encontrar el mínimo de una función f
    en el intervalo [a, b].

    Args:
        f (callable): La función objetivo a minimizar.
        a (float): Límite inferior del intervalo.
        b (float): Límite superior del intervalo.
        tol (float): Tolerancia para el criterio de parada.

    Returns:
        float: El valor de x que minimiza la función f.
    """
    phi = (1 + np.sqrt(5)) / 2  # Razón dorada
    resphi = 2 - phi

    x1 = a + resphi * (b - a)
    x2 = b - resphi * (b - a)

    f1 = f(x1)
    f2 = f(x2)

    while abs(b - a) > tol:
        if f1 < f2:
            b = x2
            x2 = x1
            f2 = f1
            x1 = a + resphi * (b - a)
            f1 = f(x1)
        else:
            a = x1
            x1 = x2
            f1 = f2
            x2 = b - resphi * (b - a)
            f2 = f(x2)

    return (a + b) / 2

def quadratic_interpolation_optimization(f, x0, h, tol=1e-5, max_iter=100):
    """
    Realiza la optimización unidimensional mediante interpolación cuadrática para encontrar
    el mínimo de una función f.

    Args:
        f (callable): La función objetivo a minimizar.
        x0 (float): Un punto inicial.
        h (float): Un tamaño de paso inicial para encontrar otros puntos.
        tol (float): Tolerancia para el criterio de parada.
        max_iter (int): Número máximo de iteraciones.

    Returns:
        float: El valor de x que minimiza la función f.
    """
    # 1. Encontrar 3 puntos iniciales x1, x2, x3 tal que x2 esté entre x1 y x3,
    # y f(x2) sea menor que f(x1) y f(x3).
    x1, x2, x3 = x0, x0 + h, x0 + 2 * h
    f1, f2, f3 = f(x1), f(x2), f(x3)

    # Asegurar que la parábola inicial tenga un mínimo
    if f1 < f2 and f2 < f3:
        # Si la función está decreciendo, mover los puntos a la izquierda
        while f1 < f2 and f2 < f3:
            x3 = x2
            f3 = f2
            x2 = x1
            f2 = f1
            x1 -= h
            f1 = f(x1)
    elif f1 > f2 and f2 > f3:
        # Si la función está creciendo, mover los puntos a la derecha
        while f1 > f2 and f2 > f3:
            x1 = x2
            f1 = f2
            x2 = x3
            f2 = f3
            x3 += h
            f3 = f(x3)
    else:
        # Si ya tenemos un mínimo entre los puntos, los dejamos como están.
        pass

    # Ordenar los puntos para que x1 < x2 < x3
    points = sorted([(x1, f1), (x2, f2), (x3, f3)], key=lambda p: p[0])
    x1, f1 = points[0]
    x2, f2 = points[1]
    x3, f3 = points[2]

    # 2. Iterar la interpolación cuadrática
    for _ in range(max_iter):
        # Calcular el vértice de la parábola que pasa por (x1, f1), (x2, f2), (x3, f3)
        # Fórmula para el mínimo de una parábola dada 3 puntos:
        # x_v = (1/2) * ( (x2^2 - x3^2)*f1 + (x3^2 - x1^2)*f2 + (x1^2 - x2^2)*f3 ) /
        #             ( (x2 - x3)*f1 + (x3 - x1)*f2 + (x1 - x2)*f3 )

        # Para evitar problemas de división por cero o inestabilidad numérica,
        # se puede usar una forma alternativa o comprobar el denominador.
        # Implementación común para el vértice de la parábola:
        num = (x2**2 - x3**2) * f1 + (x3**2 - x1**2) * f2 + (x1**2 - x2**2) * f3
        den = (x2 - x3) * f1 + (x3 - x1) * f2 + (x1 - x2) * f3

        if abs(den) < 1e-10: # Evitar división por cero
            break

        x_v = 0.5 * num / den
        f_v = f(x_v)

        # Criterio de parada
        if abs(x_v - x2) < tol or abs(f_v - f2) < tol:
            return x_v

        # Actualizar los puntos para la siguiente iteración
        if x_v < x2:
            if f_v < f2:
                x3, f3 = x2, f2
                x2, f2 = x_v, f_v
            else:
                x1, f1 = x_v, f_v
        else: # x_v > x2
            if f_v < f2:
                x1, f1 = x2, f2
                x2, f2 = x_v, f_v
            else:
                x3, f3 = x_v, f_v

        # Reordenar los puntos para mantener la convención x1 < x2 < x3
        points = sorted([(x1, f1), (x2, f2), (x3, f3)], key=lambda p: p[0])
        x1, f1 = points[0]
        x2, f2 = points[1]
        x3, f3 = points[2]

    return x2 # Retorna el mejor punto encontrado después de las iteraciones

# --- Nuevas funciones para encontrar mínimos y máximos ---

def golden_section_optimize(f, a, b, tol=1e-5, mode='min'):
    """
    Realiza la búsqueda de la razón dorada para encontrar el mínimo o máximo de una función f
    en el intervalo [a, b].

    Args:
        f (callable): La función objetivo.
        a (float): Límite inferior del intervalo.
        b (float): Límite superior del intervalo.
        tol (float): Tolerancia para el criterio de parada.
        mode (str): 'min' para minimizar, 'max' para maximizar.

    Returns:
        float: El valor de x que optimiza la función f según el modo.
    """
    if mode == 'min':
        return golden_section_search(f, a, b, tol)
    elif mode == 'max':
        # Para maximizar f(x), minimizamos -f(x)
        f_neg = lambda x: -f(x)
        result_x = golden_section_search(f_neg, a, b, tol)
        return result_x # El valor de x para el máximo de f es el mismo que para el mínimo de -f
    else:
        raise ValueError("El modo debe ser 'min' o 'max'")

def quadratic_interpolation_optimize(f, x0, h, tol=1e-5, max_iter=100, mode='min'):
    """
    Realiza la optimización unidimensional mediante interpolación cuadrática para encontrar
    el mínimo o máximo de una función f.

    Args:
        f (callable): La función objetivo.
        x0 (float): Un punto inicial.
        h (float): Un tamaño de paso inicial.
        tol (float): Tolerancia para el criterio de parada.
        max_iter (int): Número máximo de iteraciones.
        mode (str): 'min' para minimizar, 'max' para maximizar.

    Returns:
        float: El valor de x que optimiza la función f según el modo.
    """
    if mode == 'min':
        return quadratic_interpolation_optimization(f, x0, h, tol, max_iter)
    elif mode == 'max':
        # Para maximizar f(x), minimizamos -f(x)
        f_neg = lambda x: -f(x)
        result_x = quadratic_interpolation_optimization(f_neg, x0, h, tol, max_iter)
        return result_x # El valor de x para el máximo de f es el mismo que para el mínimo de -f
    else:
        raise ValueError("El modo debe ser 'min' o 'max'")

# --- Ejemplos de uso ---

# Definir una función de ejemplo para minimizar
def example_function(x):
    return x**2 - 4*x + 5 # Mínimo en x=2, f(x)=1

def another_function(x):
    return (x - 3)**2 + np.sin(x) # Una función con un mínimo no trivial

print("--- Búsqueda de la Razón Dorada (Minimización) ---")
a_golden = 0
b_golden = 5
min_x_golden = golden_section_optimize(example_function, a_golden, b_golden, mode='min')
print(f"Función: x^2 - 4x + 5")
print(f"Intervalo inicial: [{a_golden}, {b_golden}]")
print(f"Mínimo encontrado en x = {min_x_golden:.5f}")
print(f"Valor de la función en el mínimo = {example_function(min_x_golden):.5f}")

print("\n--- Búsqueda de la Razón Dorada (Maximización) ---")
# Para demostrar la maximización, usaremos una función con un máximo conocido, por ejemplo, -example_function(x)
# o simplemente la misma función y esperar el resultado correcto (x=2 para el máximo de -(x^2 - 4x + 5))
max_x_golden = golden_section_optimize(lambda x: -(x**2 - 4*x + 5), a_golden, b_golden, mode='max') # Max de -(x^2-4x+5) es Min de (x^2-4x+5) en x=2
print(f"Función: -(x^2 - 4x + 5) (buscando el máximo de x^2 - 4x + 5)")
print(f"Intervalo inicial: [{a_golden}, {b_golden}]")
print(f"Máximo encontrado en x = {max_x_golden:.5f}")
print(f"Valor de la función en el máximo = {example_function(max_x_golden):.5f}") # Usamos la función original para evaluar

print("\n--- Optimización Cuadrática (Minimización) ---")
x0_quad = 0
h_quad = 1
min_x_quad = quadratic_interpolation_optimize(example_function, x0_quad, h_quad, mode='min')
print(f"Función: x^2 - 4x + 5")
print(f"Punto inicial x0 = {x0_quad}, paso h = {h_quad}")
print(f"Mínimo encontrado en x = {min_x_quad:.5f}")
print(f"Valor de la función en el mínimo = {example_function(min_x_quad):.5f}")

print("\n--- Optimización Cuadrática (Maximización) ---")
max_x_quad = quadratic_interpolation_optimize(lambda x: -(x**2 - 4*x + 5), x0_quad, h_quad, mode='max')
print(f"Función: -(x^2 - 4x + 5) (buscando el máximo de x^2 - 4x + 5)")
print(f"Punto inicial x0 = {x0_quad}, paso h = {h_quad}")
print(f"Máximo encontrado en x = {max_x_quad:.5f}")
print(f"Valor de la función en el máximo = {example_function(max_x_quad):.5f}") # Usamos la función original para evaluar

print("\n--- Otro ejemplo con Búsqueda de la Razón Dorada (Minimización) ---")
a_golden_2 = 0
b_golden_2 = 5
min_x_golden_2 = golden_section_optimize(another_function, a_golden_2, b_golden_2, mode='min')
print(f"Función: (x - 3)^2 + sin(x)")
print(f"Intervalo inicial: [{a_golden_2}, {b_golden_2}]")
print(f"Mínimo encontrado en x = {min_x_golden_2:.5f}")
print(f"Valor de la función en el mínimo = {another_function(min_x_golden_2):.5f}")

print("\n--- Otro ejemplo con Búsqueda de la Razón Dorada (Maximización) ---")
# Para 'another_function', el máximo es más complicado de predecir visualmente.
max_x_golden_2 = golden_section_optimize(another_function, a_golden_2, b_golden_2, mode='max')
print(f"Función: (x - 3)^2 + sin(x) (buscando el máximo)")
print(f"Intervalo inicial: [{a_golden_2}, {b_golden_2}]")
print(f"Máximo encontrado en x = {max_x_golden_2:.5f}")
print(f"Valor de la función en el máximo = {another_function(max_x_golden_2):.5f}")
```
